```python
from plotly import __version__, tools
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import plotly.graph_objs as go

print(__version__) # requires version >= 1.9.0
```

```python
init_notebook_mode(connected=True)
```

```python
import numpy as np
from scipy import stats
import itertools
```

```python
samples = 10000
rv = [[1, 2], [10, 100]]
fig = tools.make_subplots(rows=2, cols=2)
colors = [['#639cd9', '#5454c5'], ['#342056', '#220e24']]
for (i, j) in itertools.product(range(2), range(2)):
    xp = np.linspace(0, 1, 1000)
    meanpdf = []
    for _ in range(samples):
        x = np.random.uniform(size=rv[i][j])
        meanpdf.append(np.mean(x))

    pdf = stats.norm.pdf(xp, np.mean(meanpdf), np.std(meanpdf))
    trace_pdf = go.Scatter(x=xp,
                           y=pdf,
                           mode='lines',
                           name='pdf',
                           line=dict(color=colors[i][j], width=3))
    hist = go.Histogram(x=meanpdf,
                        histnorm='probability density',
                        marker=dict(color=colors[i][j]),
                        opacity=0.7)

    fig.append_trace(trace_pdf, i + 1, j + 1)
    fig.append_trace(hist, i + 1, j + 1)

iplot(fig)
```

```python
samples = 10000
rv = [[1,2],[10,100]]
fig = tools.make_subplots(rows=2, cols=2)
colors = [['#e47676', '#8f1383'], ['#45056e', '#1f024c']]
for (i,j) in itertools.product(range(2),range(2)):
    xp = np.linspace(0, 1, 1000)
    meanpdf = []
    for _ in range(samples):
        x = np.random.exponential(scale=0.5, size=rv[i][j])
        meanpdf.append(np.mean(x))

    pdf = stats.norm.pdf(xp, np.mean(meanpdf), np.std(meanpdf))
    trace_pdf = go.Scatter(x=xp,
                           y=pdf,
                           mode='lines',
                           name='pdf',
                           line=dict(color=colors[i][j], width=3))
    hist = go.Histogram(x=meanpdf,
                        histnorm='probability density',
                        marker=dict(color=colors[i][j]),
                        opacity=0.7)


    fig.append_trace(trace_pdf, i+1, j+1)
    fig.append_trace(hist, i+1, j+1)
iplot(fig)
```

```python
samples = 10000
rv = [[1,2],[10,100]]
xp = np.linspace(0, 2, 1000)
fig = tools.make_subplots(rows=2, cols=2)
colors = [['#ff6d3f', '#b14b4b'], ['#594057', '#5B2333']]
for (i,j) in itertools.product(range(2),range(2)):

    meanpdf = []
    for _ in range(samples):
        x = np.random.binomial(n=2,p=0.5, size=rv[i][j])
        meanpdf.append(np.mean(x))
    pdf = stats.norm.pdf(xp, np.mean(meanpdf), np.std(meanpdf))
    trace_pdf = go.Scatter(x=xp,
                           y=pdf,
                           mode='lines',
                           name='pdf',
                           line=dict(color=colors[i][j], width=3))
    hist = go.Histogram(x=meanpdf,
                        histnorm='probability density',
                        marker=dict(color=colors[i][j]),
                        opacity=0.75)

    fig.append_trace(trace_pdf, i+1, j+1)
    fig.append_trace(hist, i+1, j+1)
iplot(fig)
```
